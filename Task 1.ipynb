```python
# Create a date range for the future weeks
future_dates = pd.date_range(start='2023-07-07', end='2023-12-31', freq='W')  
future_dates                                                                     # Adjust start and end dates as needed
```

```python
items = df["Item Number"].to_dict()
items
```

```python
# Create an empty DataFrame with columns for dates and items
data = { 'Items': [],'Date': [], 'Total Item Qty': []}
inventory_df = pd.DataFrame(data)
inventory_df
```

```python
# Populate the DataFrame with inventory levels for each item on a weekly basis
for date in future_dates:
    for item in items:
       
        inventory_level = 100 + items.index(item) * 10  # Dummy logic
        
        inventory_df = inventory_df.append({'Date': date, 'Item': item, 'Inventory_Level': inventory_level}, ignore_index=True)
```

```python
df["Inventory Receive"]
```

# Extracting month and year from the 'Date' column

```python
df['Month'] = df["Inventory Receive"].dt.strftime('%Y-%m')
df
```

# Grouping data by month and calculating total inventory for each month

```python
monthly_inventory = df.groupby('Month')['Total Item Qty'].sum().reset_index()
monthly_inventory
```


# Plotting the bar chart

```python
plt.figure(figsize=(10, 6))
plt.bar(monthly_inventory['Month'], monthly_inventory['Total Item Qty'], color='skyblue')
plt.xlabel('Month')
plt.ylabel('Total Inventory Levels')
plt.title('Total Monthly Inventory Levels')
plt.xticks(rotation=45)
plt.tight_layout()
```
